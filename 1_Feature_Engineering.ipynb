```python
# prompt: Extract the data  in following link:
# https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv

import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Read the data from the URL
df = pd.read_csv(url)

# Print the data
print(df)
```

```python
# prompt: Find number of null values in each column

null_counts = df.isnull().sum()
print(null_counts)
```

```python
# prompt: Fill missing values in column Embarked with mode value of it

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# prompt: Remove the column : Cabin

df.drop('Cabin', axis=1, inplace=True)
```

```python
# prompt: Create boxplot for column Age

import seaborn as sns

sns.boxplot(x=df['Age'])
```

```python
# prompt: Fill missing values in column Age with its median

df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python
# prompt: Remove outliers in Age column of dataset by replacing them with maximum and minimum values

# Find the maximum and minimum values of the Age column
max_age = df['Age'].max()
min_age = df['Age'].min()

# Replace outliers with maximum and minimum values
df['Age'] = df['Age'].clip(min_age, max_age)

# Print the data
print(df)
```

```python
# prompt: Create boxplot for column Age

import matplotlib.pyplot as plt

sns.boxplot(x=df['Age'])
plt.show()
```

```python
# prompt: Create an age group column by using binning

import pandas as pd
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 18, 35, 60, 100], labels=['Children', 'Young Adults', 'Adults', 'Elderly'])
df
```

```python
# prompt: Extract the data  in following link:
# https://github.com/RSRemya/Feature_Engineering/blob/main/weather_dataset.csv

import pandas as pd

url = "https://raw.githubusercontent.com/RSRemya/Feature_Engineering/main/weather_dataset.csv"

# Read the data from the URL
df = pd.read_csv(url)

# Print the data
print(df)
```

```python
# prompt: Check the datatype of each column

df.dtypes
```

```python
# prompt: Change the datatype of column : day to date and time type and then show daattypes of all columns

import pandas as pd
df['day'] = pd.to_datetime(df['day'])
print(df.dtypes)
```

```python
# prompt: Now make day column to be index column and show the dataset

df.set_index('day', inplace=True)
df
```

```python
# prompt: Remove rows having only nulls values and show the dataset

df = df.dropna(how='all')
df
```

```python
# prompt: Conduct interpolatiion through method=time to deal with missing values in the data and show the dataset

df = df.interpolate(method='time')
df
```

```python
# prompt: Fill nulls in  column event with mode of the column and show the final dataset

df['event'].fillna(df['event'].mode()[0], inplace=True)
print(df)
```
